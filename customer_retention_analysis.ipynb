Univariate analysis

```python
plt.figure(figsize=(3,5))
sns.countplot(x='Gender',data=cust)
plt.xticks(rotation=20)
plt.ylabel('Counts')
plt.xlabel('Male amd Female')
plt.title('online participation of male and female')
plt.show()
print(cust['Gender'].value_counts())
```

From the above graph we can say that more female purchased items from online when compared to men.
hence the graph for female is higher.


```python
plt.figure(figsize=(5,4))
sns.countplot(x='Age group',data=cust)
plt.ylabel('Counts')
plt.xlabel('Age range')
plt.title('category of age group')
plt.show()
print(cust['Age group'].value_counts())
```

less than 20 :1, 21-30:2, 31-40 :3, 41-50 :4, 51+:5, hence we can observe that age group 1 and 5 has very less online purchases.


```python
sns.countplot(x='City',data=cust)
plt.xticks(rotation=30)
```
