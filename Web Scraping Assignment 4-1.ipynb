#8. Scrape the details of Highest selling novels.

```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver=webdriver.Chrome(r"C:\Users\Rubia Banu\Desktop\chromedriver.exe")
driver.get('https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/')
```

```python
Book_name=[]
Author_name=[]
Volumes_sold=[]
Publisher=[]
Genre=[]
```

```python
for i in range(1):
    for j in driver.find_elements_by_xpath("//tbody/tr/td[2]"):
        Book_name.append(j.text)
    for k in driver.find_elements_by_xpath("//tbody/tr/td[3]"):
        Author_name.append(k.text)
    for l in driver.find_elements_by_xpath("//tbody/tr/td[4]"):
        Volumes_sold.append(l.text)
    for m in driver.find_elements_by_xpath("//tbody/tr/td[5]"):
        Publisher.append(m.text)
    for n in driver.find_elements_by_xpath("//tbody/tr/td[6]"):
        Genre.append(n.text)
```

```python
Novel_details=pd.DataFrame({"Book_name":Book_name[0:50],
                              "Author_name":Author_name[0:50],
                              "Volumes_sold":Volumes_sold[0:50],
                               "Publisher":Publisher[0:50],
                               "Genre":Genre[0:50]})
```

# Question 6 Scrape the details of top 100 songs on billiboard.com.

```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome(r"C:\Users\Rubia Banu\Desktop\chromedriver.exe")
driver.get('https://www.billboard.com/')
```

```python
Song_name=[]
Artist_name=[]
Lastweek_rank=[]
Peak_rank=[]
Weeks_on_board=[]
```

```python
button = driver.find_element_by_xpath("//a[@class='header__main-link header__main-link--charts']").get_attribute('href')
driver.get(button)
```

```python
button1 = driver.find_element_by_xpath("//a[@class='charts-landing__link charts-landing__video--background']").get_attribute('href')
driver.get(button1)
```

```python
start=0
end=1
for page in range(start,end):
    song=driver.find_elements_by_xpath("//li[@class='chart-list__element display--flex']/button/span[2]/span[1]")
    for i in song:
        Song_name.append(i.text)
    
    artist=driver.find_elements_by_xpath("//li[@class='chart-list__element display--flex']/button/span[2]/span[2]")
    for i in artist:
        Artist_name.append(i.text)
    
    week=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']")
    for i in week:
        Lastweek_rank.append(i.text)
    
    peak=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']")
    for i in peak:
        Peak_rank.append(i.text)
        
    weekboard=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']")
    for i in weekboard:
        Weeks_on_board.append(i.text)

        
    
```

```python
details=pd.DataFrame({"Song_name":Song_name[0:100],
                      "Artist_name":Artist_name[0:100],
                      "Lastweek_rank":Lastweek_rank[0:100],
                      "Peak_rank":Peak_rank[0:100],
                      "Weeks_on_board":Weeks_on_board[0:100]})
```

```python
details
```

# Question 2 Scrape the details team India’s international fixtures from bcci.tv

```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome(r"C:\Users\Rubia Banu\Desktop\chromedriver.exe")
driver.get('https://www.bcci.tv/')
```

```python
path = driver.find_element_by_xpath("//a[@class='navigation__link navigation__link--in-drop-down']").get_attribute('href')
driver.get(path)
```

```python
Match_title=[]
Series=[]
Place=[]
Date=[]
Time=[]
```

```python
start=0
end=1
for page in range(start,end):
    match=driver.find_elements_by_xpath("//div[@class='fixture__teams']")
    for i in match:
        Match_title.append(i.text)
    
    series=driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__tournament-label u-truncated']")
    for i in series:
        Series.append(i.text)
    
    place=driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span")
    for i in place:
        Place.append(i.text)
    
    date=driver.find_elements_by_xpath("//div[@class='fixture__full-date']/span")
    for i in date:
        Date.append(i.text)
        
    time=driver.find_elements_by_xpath("//div[@class='fixture__date-details']/span[2]")
    for i in time:
        Time.append(i.text)
```

```python
Match_details=pd.DataFrame({"Match_title":Match_title[0:],
                      "Series":Series[0:],
                      "Place":Place[0:],
                      "Date":Date[0:],
                      "Time":Time[0:]})
```
