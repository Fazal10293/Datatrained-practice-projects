```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver=webdriver.Chrome(r"C:\Users\Rubia Banu\Desktop\chromedriver.exe")
driver.get("https://www.imdb.com/list/ls095964455/")
```

```python
Name=[]
Year_span=[]
Genre=[]
Run_time=[]
Ratings=[]
Votes=[]
```

```python
start=0
end=1
for page in range(start,end):
    name=driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a")
    for i in name:
        Name.append(i.text)
    
    yr=driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']")
    for i in yr:
        Year_span.append(i.text)
    
    gen=driver.find_elements_by_xpath("//span[@class='genre']")
    for i in gen:
        Genre.append(i.text)
    
    run=driver.find_elements_by_xpath("//span[@class='runtime']")
    for i in run:
        Run_time.append(i.text)
        
    rat=driver.find_elements_by_xpath("//div[@class='ipl-rating-widget']/div[1]/span[2]")
    for i in rat:
        Ratings.append(i.text)

        
    vot=driver.find_elements_by_xpath("//span[@name='nv']")
    for i in vot:
        Votes.append(i.text)

        
    
    
    
```

```python
df=pd.DataFrame({'Name':Name[0:100],
                'Year':Year_span[0:100],
                'Genre':Genre[0:100],
                 'Ratings':Ratings[0:100],
                'Run_time':Run_time[0:100],
                'Votes':Votes[0:100]})
```
